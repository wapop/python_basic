```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# 읽어드릴 파일 경로 : /dataset/bike_sharing_demand_train.csv
filepath = "./datasets/exercise/bike_sharing_demand_train.csv"
data = pd.read_csv(filepath)
data
```

#### 날짜 데이터 파싱

```python
year = []
month = []
day = []

for i in range(len(data)):
    year.append(data.datetime[i][:4])
    month.append(data.datetime[i][5:7])
    day.append(data.datetime[i][8:10])
```

```python
data['year'] = year
data['month'] = month
data['day'] = day
```

#### 소수점 자리수 변경

```python
data.atemp = data.atemp.apply(lambda x : '%.2f' % x)
```

```python
data.atemp
```

```python
data.atemp = data.atemp.astype('float64')
```

#### 이상치 처리

```python
data.isna().sum()
```

```python
data.describe()
```

```python
data[data['windspeed']>=50]
```

```python
data['windspeed'] = data['windspeed'].apply(lambda x: 50 if x>=50 else x) # windspeed가 50 이상인 경우에 값 변경하기
```

#### 계절, 날씨 컬럼 추가

```python
se = pd.DataFrame({'season_val': ['봄','여름','가을','겨울']}, index=[1,2,3,4])
se
```

```python
we = pd.DataFrame({'weather_val': ['맑은 날씨','약간의 안개와 구름','약간의 눈과 비','아주 많은 비, 우박']}, index=[1,2,3,4])
we
```

```python
data = pd.merge(data, se, left_on='season', right_index=True)
data.head()
```

```python
data = pd.merge(data, we, left_on='weather', right_index=True)
data.head()
```

```python
b_count = [0,50,200,500,800,1000]
b_names = ['매우 적음', '적음', '보통', '많음', '매우 많음']
```

```python
cats= pd.cut(data['count'], b_count, labels=b_names)
```
