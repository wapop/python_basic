```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# 읽어드릴 파일 경로 : /bike_sharing_demand_train.csv
filepath = "./bike_sharing_demand_train.csv"
df = pd.read_csv(filepath)
df
```

```python
df['season'] = df['season'].replace([1,2,3,4],['봄','여름','가을','겨울'])
df['weather'] = df['weather'].replace([1,2,3,4],['맑음','흐림','비','눈'])
```

```python
df['holiday'] = df['holiday'].replace([0,1],['예','아니오'])
df['workingday'] = df['workingday'].replace([0,1],['예','아니오'])
```

```python
df.columns = ['날짜','계절','휴일유무','출근길유무','날씨','온도','체감온도','습도','풍속','비회원대여','회원대여','합계대여']
```

```python
df.describe()
```

```python
format = lambda x : '%.2f 도' % x 
df['체감온도'] = df['체감온도'].map(format)
```

```python
format = lambda x :'%.2f 도' % x 
df['온도'] = df['온도'].map(format)
df
```

```python
bin = [0,50,150,250,300,350]
```

```python
cats = pd.cut(df['합계대여'],bin)
cats
```

```python
cats
```

```python
group_name = ['매우적음','적음','보통','많음','매우많음']
```

```python
cate = pd.cut(df['합계대여'],bin, labels=group_name)
```

```python
df['카테고리'] = cate
df
```

```python
df['날짜'] = df['날짜'].astype("datetime64")
```

```python
df['연도'] = df['날짜'].dt.year
df['월'] = df['날짜'].dt.month
df['일'] = df['날짜'].dt.day
df['시간'] = df['날짜'].dt.hour
```

```python
df
```

```python
df.to_csv('10조_신민철.csv',index=False ,encoding='utf-8-sig')
```
