```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
# 읽어드릴 파일 경로 : /dataset/bike_sharing_demand_train.csv
filepath = "./datasets/exercise/bike_sharing_demand_train.csv"
data = pd.read_csv(filepath)
data
```

### 날짜 데이터 파싱하기

```python
type('datetime')
```

```python
data['datetime'] = data['datetime'].astype("datetime64")
```

```python
data['year'] = data['datetime'].dt.year
data['month'] = data['datetime'].dt.month
data['day'] = data['datetime'].dt.day
data['hour'] = data['datetime'].dt.hour
```

```python
data.head()
```

### 자전거 대여횟수를 카테고리화 하기(예: 매우 적음, 적음, 보통, 많음, 매우 많음)

```python
count_rg = [0,100, 300, 500, 700, 1000]
names_rg = ['매우 적음', '적음', '보통', '많음', '매우 많음']
data_count = pd.cut(data['count'], count_rg,labels = names_rg)
```

```python
data_count
```

```python
data['data_count'] = data_count
```

```python
data.head()
```

### 컬럼명 변경하기

```python
data = data.rename(columns={'datetime':'시간정보',
                            'season':'계절',
                            'holiday':'공휴일',
                            'workingday': '근무일',
                            'weather': '날씨',
                            'temp': '기온',
                            'atemp': '체감기온',
                            'humidity': '상대습도',
                            'windspeed': '풍속',
                            'casual': '미등록사용자',
                            'registered': '등록사용자',
                            'count': '대여수량',
                            'year': '연도',
                            'month': '월',
                            'day': '일',
                            'hour': '시간대',
                            'data_count': '대여량'})
data.head()
```
