```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
base_url = 'https://movie.naver.com/'
seoul_url = 'movie/sdb/rank/rmovie.nhn?sel=pnt&date=20200720'
```

```python
response = requests.get(base_url+seoul_url)
response
```

```python
response.text
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
lists = soup.findAll('div', {'class' : 'tit5'}) 
lists
```

```python
for item in lists:
    name= item.find('a')
    print(name.text)
    movie_name_list.append(name.text)
```

```python
score_lists = soup.findAll('td', {'class' : 'point'}) 
score_lists
```

```python
movie_score_list = []
movie_name_list = []
movie_url_list = []
```

```python
for item in lists:
    # 링크 가져오기
    hrefs = item.find('a')
    href = hrefs.get('href')
    full_url = base_url + href
    movie_score_list.append(score_lists)

    movie_url_list.append(full_url)
```

```python
print(len(movie_score_list))
print(len(movie_name_list))
print(len(movie_url_list))
```

```python
result = pd.DataFrame(data={'[Score]' : movie_score_list, '[NAME]' : movie_name_list, '[URL]' : movie_url_list})
print(result)
result
```
