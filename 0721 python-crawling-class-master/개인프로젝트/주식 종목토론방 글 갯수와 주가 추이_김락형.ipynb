```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
import os
```

```python
# 크롤링할 웹 사이트 정보
base_url = 'https://finance.naver.com' # 크롤링할 사이트 ( 기본  url)
my_url = '/item/board.nhn?code=020150&page='
page_num= '&page=' #

```

```python
# get html

## 크롬창을 직접 
options = Options()
# options.headless = True //// 이 구문을 실행시키면 window 화면에 실행 없이 Background에서 실행된다.
browser = webdriver.Chrome(options=options) # chrome web driver를 같은 폴더에 넣어 주어야함
# 직접 실행시키고 
browser.implicitly_wait(1) #1초 휴식

#지정한 url로 접속
browser.get(base_url+my_url)
time.sleep(1)  ## 1초 휴식

# BeautifulSoup을 활용하여 데이터 파싱
html = browser.page_source
soup = BeautifulSoup(html, 'html.parser')
```
