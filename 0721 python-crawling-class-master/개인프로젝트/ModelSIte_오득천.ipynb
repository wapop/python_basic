```python
# (실습3) 이미지
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver


from selenium.webdriver.chrome.options import Options
import time
import os
```

```python
# 크롤링할 웹 사이트 정보

class ModelSite:
    base_url = 'http://www.findmodel.co.kr'
    startPageNum = 1
    modolList = []


    def getModelURL(self,pageIndex):
        pageNum = str(startPageNum + pageIndex)
        addUrl = f'/ModelFinder?PageIndex={pageNum}&CategoryId=0&Gender=All&MinimumAge=0&MaxAge=0&MinimumHeight=0&MaxHeight=0&MinimumWeight=0&MaxWeight=0&SidoId=0&GugunId=0&Career=ALL&OrderType=0&IsDetailSearch=False'
        return addUrl
    
    def getHTML(self,pageIndex):
        # get html
        #사진을 얻을려면 그냥 request로는 안된다. html을 받고나서 javascript가 이미지를 다운 받기 때문이다.
        ## 크롬창을 직접
        options = Options()
        # options.headless = True              #크롬이 명시적으로 열리지 않고 BackGround에서 실행되고 싶을때
        browser = webdriver.Chrome(options=options)
        browser.implicitly_wait(3)
        browser.get(base_url+getModelURL(pageIndex))
        time.sleep(5)
        # BeautifulSoup을 활용하여 데이터 파싱
        html = browser.page_source
        soup = BeautifulSoup(html, 'html.parser')

        # 이미지
        photos = soup.findAll('a', {'class' : 'model-box'}) #selenium으로 실행시켜야 javascript가 제대로 동작해서 image 태그가 달린다.
        print(photos)


        img_dir = f"{os.getcwd()}/img_model"
        if not os.path.isdir(img_dir):
           os.mkdir(img_dir)

        for photo in photos :
            modelInfo = {'name':'','img':'','age':'','tall':''}
            modelInfo['img'] = base_url+photo.find('img')['src']
            print(modelInfo['img'])
            modelInfo['name'] = photo.find('strong').text
            modelInfo['age'] = photo.find('p',{'class' : 'info'}).find('span').text.split(' ')[0].split('세')[0]
            print(modelInfo['age'])
            modelInfo['tall'] = photo.select('p > span')[0].text.split(' ')[2]
            print(modelInfo['tall'])
            modolList.append(modelInfo)

        for model in modolList:
            img_data = requests.get(model['img']).content
            name = model['name']
            with open(f'img_model/{name}.png', 'wb') as handler:
                handler.write(img_data)

    def saveImage(self,endPageNum):
        for i in range(EndPageNum):
            modelSite.getHTML(i) 

modelSite = ModelSite()
modelSite.saveImage(2)
```
