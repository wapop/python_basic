```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
import os

```

```python
# 크롤링할 웹 사이트 정보
base_url = 'http://www.buzzbee.co.kr'
page_url = '/shop/goods/goods_list.php?category=001001022'

#http://www.buzzbee.co.kr/shop/goods/goods_list.php?category=001001022

# get html
response = requests.get(base_url+page_url)
# response.text
```

```python
# 잘 받아졌나 확인하기

response.text
```

```python
# BeautifulSoup을 활용하여 데이터 파싱
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
# a태그에 스타일 속성중에 요런거 찾아줘~

lists = soup.findAll('a',{'style':'color:#888; font-size:11px;'})
lists
```

```python
#기타이름 받아보기

guitar_name_list = []
guitar_price_list = []
for item in lists:
    
    name = item.text.split('] ')
    
    if len(name) == 2:
        guitar_name = name[1]
        guitar_name_list.append(guitar_name)
    
guitar_name_list
```

```python
# 기타 가격 뽑아보기
price_lists = soup.findAll('b',{'style':'color:#4c78bc'})
price_lists
```

```python
# 기타 가격 뽑아보기
guitar_price_list = []
for item in price_lists:
    price = item.text.strip('\\')
    guitar_price_list.append(price)
    
guitar_price_list
```

```python
# URL 리스트 뽑아보기
guitar_url_list = []
for item in lists:
    href = item.get('href')
    full_url = base_url+href
    guitar_url_list.append(full_url)
guitar_url_list
```

```python
df = pd.DataFrame(data={'Name' : guitar_name_list, 'Price' : guitar_price_list, 'Url' : guitar_url_list})
df
```

```python
df.info()
```

```python
df['Brand'] = df['Name'].str.slice(start=0, stop=6)
df
```

```python
# price에서 ','없애기

df['Price'] = df['Price'].apply(lambda x: x.replace(',', ''))
df
```

```python
# type int형으로 바꾸기

df['Price'] = pd.to_numeric(df['Price'])
df.info()
```
