```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
base_url = 'https://www.tripadvisor.co.kr'
ilsan_url = '/RestaurantsNear-g2024865-d1483296-Ilsan_Lake_Park-Goyang_Gyeonggi_do.html'
```

```python
response = requests.get(base_url+ilsan_url)
response
```

```python
response.text
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
lists = soup.findAll('div', {'class' : 'location_name'})
lists
```

```python
for item in lists:
    name = item.text
    print(name)
```

```python
# 데이터 파싱하기

rest_name_list = []
for item in lists:
    name = item.text.split(' ')
    rest_name_list.append(name[0])
print(rest_name_list)
```

```python
# 데이터 프레임을 만들기 위한 리스트
        
#  print('[ID]' + rest_id + ' [NAME]' + rest_name + ' [URL]' + rest_url)
        
```

```python
# 데이터 파싱하기
rest_url_list = []

for item in lists:
    # 링크 가져오기
    hrefs = item.find('a')
    href = hrefs.get('href')
#     print(href)
    full_url = base_url + href
    rest_url = full_url
    # 순위와 이름 분리하기, 광고 제거하기
    #[1 , 장생건강원]
    #[2,  무슨음식점]
    #[광고음식점]
    rest_url_list.append(rest_url)
```

```python
rest_url_list
```

```python
print(len(rest_url_list))
print(len(rest_name_list))
```

```python
ID = list(range(30))
ID
```

```python
ILSAN_Restaurant = pd.DataFrame(data={'[ID]' : ID, '[NAME]' : rest_name_list, '[URL]' : rest_url_list})
ILSAN_Restaurant.head()
```

```python
ILSAN_Restaurant.to_csv("ilsan_restaurant.csv")
```
