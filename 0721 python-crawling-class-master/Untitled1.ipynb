```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
import os
```

```python
base_url = 'https://movie.naver.com/'
seoul_url = 'movie/running/current.nhn'
```

```python
options = Options()
```

```python
browser = webdriver.Chrome(options=options)
browser.implicitly_wait(3)
browser.get(base_url+seoul_url)
time.sleep(5)
```

```python
html = browser.page_source
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
photos = soup.findAll('div', {'class' : 'thumb'})
print(photos)
```

```python
img_di = f"{os.getcwd()}/img"
if not os.path.isdir(img_di):
   os.mkdir(img_di)
```

```python
for photo in photos :
    url = photo.get('style')[1]
    img_file_name = url.split('/')[-1]
    img_data = requests.get(url).content
    with open(f'img/i{img_file_name}', 'wb') as handler:
        handler.write(img_data)
```
