```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 읽어드릴 파일 경로 : /dataset/bike_sharing_demand_train.csv
filepath = "./dataset/bike_sharing_demand_train.csv"
data = pd.read_csv(filepath)
```

```python
data['datetime']=data['datetime'].astype('datetime64')
data['year']=data['datetime'].dt.year
data['month']=data['datetime'].dt.month
data['day']=data['datetime'].dt.day
data['hour']=data['datetime'].dt.hour
data['minute']=data['datetime'].dt.minute
data['second']=data['datetime'].dt.second
```

```python
data["season"] = data.season.map({1: "Spring", 2 : "Summer", 3 : "Fall", 4 :"Winter" })
plt.figure(figsize=(10,6))

plt.title("Season & count Relation")
sns.barplot(x="season", y="count", data=data)

g = plt.ylabel("Bike sharing demand (Count)") 
z = plt.xlabel("Season")
```
