```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 읽어드릴 파일 경로 : /dataset/bike_sharing_demand_train.csv
filepath = "./dataset/bike_sharing_demand_train.csv"
data = pd.read_csv(filepath)
```

```python
data.head()
```

```python
# 4종류 이상 그리기, 의미 도출
```

```python
plt.figure(figsize=(12, 3))
sns.pointplot(data=data, x="season", y="count")
```

### 가을 시즌에 대여량이 가장 많다.

```python
plt.figure(figsize=(26, 5))
sns.barplot(data=data, x="temp", y="count")
```

### 온도가 28~37도 구간일 때 대여량이 많다. 

```python
plt.figure(figsize=(26, 5))
sns.lineplot(data=data, x="humidity", y="count")
```

### 습도 20~40 구간 이용이 많고 높을수록 대여량이 줄어든다.

```python
data['datetime']=data['datetime'].astype('datetime64')
```

```python
data['year']=data['datetime'].dt.year
data['month']=data['datetime'].dt.month
data['day']=data['datetime'].dt.day
data['hour']=data['datetime'].dt.hour
data['minute']=data['datetime'].dt.minute
data['second']=data['datetime'].dt.second
```

```python
data.head()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=data, x="year", y="registered")
```

### 2012년도에 회원제 운영 비율이 높다

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=data, x="month", y="casual")
```

### 대여량과 비회원제 비율은 정관계에 있다.

```python
plt.figure(figsize=(8, 5))
sns.barplot(data=data, x="month", y="registered")
```

### 대여량과 회원제 비율은 정관계에 있으며, 비회원제 비율보다 회원제 비율이 높다

```python
plt.figure(figsize=(12, 3))
sns.swarmplot(data=data, x="month", y="count")
```

### 9월 이용량이 가장 높고, 1월이 가장 낮다.

```python
sns.distplot(data["count"])
```

### 이용량은 0 ~ 1,000 구간이며, 0~200 구간이 가장 높다.
