```python
filepath_data['datetime'] = filepath_data['datetime'].astype('datetime64')
```

```python
filepath_data.dtypes
```

```python
filepath_data['year'] = filepath_data['datetime'].dt.year
filepath_data['month'] = filepath_data['datetime'].dt.month
filepath_data['day'] = filepath_data['datetime'].dt.day
filepath_data['hour'] = filepath_data['datetime'].dt.hour
filepath_data['minute'] = filepath_data['datetime'].dt.minute
filepath_data['second'] = filepath_data['datetime'].dt.second
filepath_data.head()
```

```python
sns.barplot(x= 'year', y ='count', data = filepath_data)
```
