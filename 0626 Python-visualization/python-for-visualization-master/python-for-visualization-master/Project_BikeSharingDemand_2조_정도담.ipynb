```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 읽어드릴 파일 경로 : /dataset/bike_sharing_demand_train.csv
filepath = "./dataset/bike_sharing_demand_train.csv"
data = pd.read_csv(filepath)
```

```python
data.head()
```

```python
data['datetime']=data['datetime'].astype('datetime64')
```

```python
data['year']=data['datetime'].dt.year
data['month']=data['datetime'].dt.month
data['day']=data['datetime'].dt.day
data['hour']=data['datetime'].dt.hour
data['minute']=data['datetime'].dt.minute
data['second']=data['datetime'].dt.second
```

###가설 설정
#1-1, #1-2 : 
가설: 봄보다 겨울의 이용량이 높은 이유로 워킹데이 혹은 휴일 유무가 영향을 미칠 것이다.
#2 휴일에는 체감기온이 높아도 이용량이 많을 것이다. 
#3 비회원제일때 이용량은 날씨와 영향이 있을 것이다.
#4 워킹데이에는 시간대별 이용량이 출퇴근시간에 많을 것이다.


```python
data["season"] = data.season.map({1: "Spring", 2 : "Summer", 3 : "Fall", 4 :"Winter" })
plt.figure(figsize=(10,6))

plt.title("Season & count Relation")
sns.barplot(x="season", y="count", data=data)

g = plt.ylabel("Bike sharing demand (Count)") 
z = plt.xlabel("Season")
```

```python
#1-1 계절별 평일/휴일 간 이용량 차이(평일)
sns.barplot(x="season",y="count", hue="workingday", data=data)
```

```python
#1-2 계절별로 평일/휴일 간 이용량 차이(주말)
sns.barplot(x="season",y="count", hue="holiday", data=data)
```

--> 봄, 가을 겨울에는 평일에 더 많이 대여하는데 여름에는 주말에 더 많이 대여한다.

```python
#2 체감기온과 휴일의 상관관계
sns.relplot(x="atemp",y="count",
            size="holiday",
            kind="line",data=data)
```

--> 체감기온이 높을수록 휴일 유무에 따른 이용량 차이가 크다.

```python
#3 비회원제 사용량과 날씨의 상관관계
sns.pointplot(data=data, x="weather", y="casual")
```

--> 날씨가 좋을 때 비회원들의 사용량이 많다.

```python
#4 워킹데이에는 시간대별 이용량이 출퇴근시간에 많을 것이다.
ax=sns.scatterplot(x="datetime", y="workingday",data=data)
```
