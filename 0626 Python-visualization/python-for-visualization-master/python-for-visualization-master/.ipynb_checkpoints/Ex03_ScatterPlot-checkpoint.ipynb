# Scatter Plot 

## 산포도 &  버블차트

https://seaborn.pydata.org/generated/seaborn.scatterplot.html

```python
import seaborn as sns; sns.set()
import matplotlib.pyplot as plt
tips = sns.load_dataset("tips")
ax = sns.scatterplot(x="total_bill", y="tip", data=tips)
```

```python
ax = sns.scatterplot(x="total_bill", y="tip", hue="time", data=tips)
```

```python
ax = sns.scatterplot(x="total_bill", y="tip", hue="time", style="time", data=tips)
```

```python
ax = sns.scatterplot(x="total_bill", y="tip", hue="day", style="time", data=tips)
```

```python
ax = sns.scatterplot(x="total_bill", y="tip", size="size", data=tips)
```

```python
ax = sns.scatterplot(x="total_bill", y="tip", hue="size", size="size", data=tips)
```

```python
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
ax = sns.scatterplot(x="total_bill", y="tip",
                     hue="size", size="size",
                     palette=cmap,
                     data=tips)
```

```python
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
ax = sns.scatterplot(x="total_bill", y="tip",
                     hue="size", size="size",
                     sizes=(20, 200), palette=cmap,
                     legend="full", data=tips)
```

```python
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
ax = sns.scatterplot(x="total_bill", y="tip",
                     hue="size", size="size",
                     sizes=(20, 200), hue_norm=(0, 7),
                     legend="full", data=tips)
```

```python
cmap = sns.cubehelix_palette(dark=.3, light=.8, as_cmap=True)
ax = sns.scatterplot(x="total_bill", y="tip",
                     hue="day", size="smoker",
                     palette="Set2",
                     data=tips)
```

```python
markers = {"Lunch": "s", "Dinner": "X"}
ax = sns.scatterplot(x="total_bill", y="tip", style="time",
                     markers=markers,
                     data=tips)
```
