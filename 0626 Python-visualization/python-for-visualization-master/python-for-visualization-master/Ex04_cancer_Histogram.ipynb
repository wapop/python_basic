```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# 읽어드릴 파일 경로 지정
cancer_b_filepath = "./dataset/cancer_b.csv"
cancer_m_filepath = "./dataset/cancer_m.csv"

# 변수에 파일에 있는 데이터 셋 로드
cancer_b_data = pd.read_csv(cancer_b_filepath, index_col="Id")
cancer_m_data = pd.read_csv(cancer_m_filepath, index_col="Id")

# 데이터 출력 
cancer_b_data.head()
```

```python
cancer_b_data.head()
```

```python
sns.distplot(a=cancer_b_data['Area (mean)'], label="B")
sns.distplot(a=cancer_m_data['Area (mean)'], label="M")
plt.legend()
```

```python
# Histograms for benign and maligant tumors
sns.distplot(a=cancer_b_data['Area (mean)'], label="Benign", kde=False)
sns.distplot(a=cancer_m_data['Area (mean)'], label="Malignant", kde=False)
plt.legend()
```

```python
sns.distplot(a=cancer_b_data['Radius (worst)'], label="Benign", kde=False)
sns.distplot(a=cancer_m_data['Radius (worst)'], label="Malignant", kde=False)
plt.legend()
```

```python
sns.kdeplot(data=cancer_b_data['Radius (worst)'], shade=True, label="Benign")
sns.kdeplot(data=cancer_m_data['Radius (worst)'], shade=True, label="Malignant")
```
