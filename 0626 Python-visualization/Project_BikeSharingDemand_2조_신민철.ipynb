```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 읽어드릴 파일 경로 : /dataset/bike_sharing_demand_train.csv
filepath = "./dataset/bike_sharing_demand_train.csv"
data = pd.read_csv(filepath)
```

```python
data.head()
```

```python
data.dtypes
```

```python
data['datetime']=data['datetime'].astype('datetime64')
```

```python
data['year']=data['datetime'].dt.year
data['month']=data['datetime'].dt.month
data['day']=data['datetime'].dt.day
data['hour']=data['datetime'].dt.hour
data['minute']=data['datetime'].dt.minute
data['second']=data['datetime'].dt.second
```

```python
plt.figure(figsize=(10,6))

plt.title("Hour & Count Relation")
sns.barplot(x="hour", y="count", data=data)

g = plt.ylabel("Bike sharing demand (Count)") 
z = plt.xlabel("Hour")
```

```python
data["season"] = data.season.map({1: "Spring", 2 : "Summer", 3 : "Fall", 4 :"Winter" })
```

```python
plt.figure(figsize=(10,6))

plt.title("Season & count Relation")
sns.barplot(x="season", y="count", data=data)

g = plt.ylabel("Bike sharing demand (Count)") 
z = plt.xlabel("Season")
```

```python
plt.figure(figsize=(100,20))

plt.title("temp & Count Relation")
sns.barplot(x="temp", y="count", data=data)

g = plt.ylabel("Bike sharing demand (Count)") 
z = plt.xlabel("temp")
```

```python
plt.figure(figsize=(5,10))

plt.title("Atemp & Count Relation")
sns.lineplot(x="atemp", y="count", data=data, ci=None)

g = plt.ylabel("Bike sharing demand (Count)") 
z = plt.xlabel("atemp")
```
